```python
"""Add lot names to enforcement data for easier lookups"""import pandas as pdprint("Loading data...")enforcement = pd.read_csv('../data/processed/enforcement_lot_level_extended.csv', parse_dates=['datetime'])lot_mapping = pd.read_csv('../data/lot_mapping_enhanced.csv')print(f"Enforcement records: {len(enforcement):,}")print(f"Lot mapping: {len(lot_mapping):,} lots")# Create lot number -> name mappinglot_names = lot_mapping[['Lot_number', 'alternative_location_description', 'Zone_Name']].drop_duplicates('Lot_number')# Merge to add namesenforcement = enforcement.merge(    lot_names,    on='Lot_number',    how='left')print(f"\nMerged enforcement data:")print(f"  Records: {len(enforcement):,}")print(f"  Lots with names: {enforcement['alternative_location_description'].notna().sum():,}")# Fill missing alternative_location_description with Zone_Name as fallbackenforcement['Lot_Name'] = enforcement['alternative_location_description'].fillna(enforcement['Zone_Name'])# Saveoutput_file = '../data/processed/enforcement_lot_level_extended.csv'enforcement.to_csv(output_file, index=False)print(f"\nSaved: {output_file}")print(f"\nSample lot names:")print(enforcement[['Lot_number', 'Lot_Name']].drop_duplicates('Lot_number').head(20))
```
